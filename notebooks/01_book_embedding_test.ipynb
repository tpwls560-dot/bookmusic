```python
from sentence_transformers import SentenceTransformer
import numpy as np
import pandas as pd
```

```python
#샘플 데이터
books=[
    {'id': 'B001', 'title': '해리포터', 'summary':'마법 학교에서 펼쳐지는 모험 이야기'},
    {'id': 'B002', 'title': '반지의 제왕', 'summary':'절대 반지를 둘러싼 위대한 판타지 전쟁'},
    {'id': 'B003', 'title': '트와일라잇', 'summary': '인간과 뱀파이어의 금지된 사랑 이야기'}
]
df_books=pd.DataFrame(books)
```

```python
#SBERT 모델 로딩
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

```python
#책 summary를 Embedding
book_embeddings=model.encode(df_books['summary'].tolist())
```

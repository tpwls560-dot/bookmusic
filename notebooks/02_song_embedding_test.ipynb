```python
from sentence_transformers import SentenceTransformer
import numpy as np
import pandas as pd
```

```python
# 샘플 데이터
songs = [
    {"id": "S101", "lyrics": "magic mystical wizard sorcery"},
    {"id": "S102", "lyrics": "love story between two people deeply in love"},
    {"id": "S103", "lyrics": "epic battle with swords and dragons"}
]
df_songs=pd.DataFrame(songs)
```

```python
# SBERT 모델 로딩
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

```python
# 노래 Summary를 Embedding
song_embeddings = model.encode(df_songs["lyrics"].tolist())

print("Embedding shape:", song_embeddings.shape)
song_embeddings[:2]
```

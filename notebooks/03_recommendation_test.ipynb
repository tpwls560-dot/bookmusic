```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd
```

```python
books=[
    {'id': 'B001', 'summary': '마법 학교에서 펼쳐지는 모험 이야기'},
    {'id': 'B002', 'summary': '절대 반지를 둘러싼 판타지 전쟁 대서사시'}
]
songs=[
    {"id": "S101", "lyrics": "magic mystical wizard sorcery"},
    {"id": "S102", "lyrics": "love story between two people deeply in love"},
    {"id": "S103", "lyrics": "epic battle with swords and dragons"}
]

df_books=pd.DataFrame(books)
df_songs=pd.DataFrame(songs)

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

book_emb=model.encode(df_books['summary'])
song_enb=model.encode(df_songs['lyrics'])
```

```python
#책 B001에 대해 노래 추천
target_book_emb=book_emb[0].reshape(1,-1)

similarities=cosine_similarity(target_book_emb, song_enb)[0]

df_songs['score']=similarities

recommend=df_songs.sort_values(by='score',ascending=False)
recommend
```
